### Q.2. In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.

### dtypes

### to_datetime

```python
data.head(2)
```

```python
data['Release_Date'] = data['Release_Date'].str.strip()  # Strip leading/trailing spaces
data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce')

# 'errors='coerce'' will set invalid parsing attempts to NaT (Not a Time)
```

```python
data.dtypes
```
